```python
eda_titanic
```

### Handling null values in Age column 
#### there are 177 null values
#### Age is numeric hence could be filled by mean,median or mode depending on the distribution

```python
eda_titanic.hist(column=['Age'], bins=6)
```

#### Since the above distribution is right skewed we can use median to fillin the null values

```python
 eda_titanic['Age']=eda_titanic['Age'].fillna(value=eda_titanic['Age'].median()) 
```

```python
eda_titanic
```

### Handling null values in Embarked column 
#### there are 2 null values
#### Embarked is categorical hence could be filled by mode
#### Since by the observations we know that S(Southampton) appears frequently, null values in Embarked column can be filled with S

```python
eda_titanic=eda_titanic.fillna("S")
```

## 5. Outliers
#### Outliers can be detected using boxplots or scatterplots
#### Outliers found in Age,Fare,SibSp,Parch

#### 1. Handling outliers in Age

```python
eda_titanic.boxplot(column = ['Age'])
```

#### We can observe that outliers appear both at top and bottom

```python
 eda_titanic['Age'].describe()
```

```python
IQR = eda_titanic['Age'].quantile(0.75)-eda_titanic['Age'].quantile(0.25)
print(IQR)
```

```python
Upper_OutlierLimit = eda_titanic['Age'].quantile(0.75) + 1.5*IQR 
Lower_OutlierLimit = eda_titanic['Age'].quantile(0.25) - 1.5*IQR 
print(Upper_OutlierLimit) 
print(Lower_OutlierLimit)
```

```python
OutlierValues_age = eda_titanic[(eda_titanic['Age']>=Upper_OutlierLimit) | (eda_titanic['Age']<=Lower_OutlierLimit)]
```

```python
OutlierValues_age
```

```python
OutlierValues_age.reset_index(inplace=True)
```

```python
OutlierValues_age
```

#### These outliers are capped

```python
for i in (OutlierValues_age['index']):
    if (eda_titanic['Age'][i]>=Upper_OutlierLimit):
        eda_titanic['Age'][i]=eda_titanic['Age'].quantile(0.95)
    else:
        eda_titanic['Age'][i]=eda_titanic['Age'].quantile(0.05) 
        
```

```python
obj = eda_titanic.dtypes == np.object 
print(obj)
```

```python
eda_titanic.columns[obj]
```

```python
dummydf = pd.DataFrame()
for i in eda_titanic.columns[obj]:     
    dummy=pd.get_dummies(eda_titanic[i], drop_first=True)                  
    dummydf=pd.concat([dummydf, dummy], axis=1) 
print(dummydf)
```
